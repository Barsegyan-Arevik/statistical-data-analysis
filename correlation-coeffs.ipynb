<h3>1. Коэффициент корреляции Пирсона</h3>

```python
def calculate_pearson_correlation(data, variable1, variable2):
    n = len(data)
    
    # Вычисление средних значений
    mean1 = sum(variable1) / n
    mean2 = sum(variable2) / n
    
    # Вычисление суммы произведений отклонений
    sum_deviation_product = sum((x - mean1) * (y - mean2) for x, y in zip(variable1, variable2))
    
    # Вычисление суммы квадратов отклонений
    sum_squared_deviation1 = sum((x - mean1) ** 2 for x in variable1)
    sum_squared_deviation2 = sum((y - mean2) ** 2 for y in variable2)
    
    # Вычисление ковариации и стандартных отклонений
    covariance = sum_deviation_product / (n - 1)
    std_deviation1 = (sum_squared_deviation1 / (n - 1)) ** 0.5
    std_deviation2 = (sum_squared_deviation2 / (n - 1)) ** 0.5
    
    # Вычисление коэффициента корреляции Пирсона
    correlation = covariance / (std_deviation1 * std_deviation2)
    
    return correlation


# Расчет коэффициента корреляции Пирсона для каждой пары переменных
for combination in combinations:
    var1 = combination[0]
    var2 = combination[1]
    var3 = combination[2]
    
    # Получение данных для соответствующих переменных
    var = []
    var.append(var1)
    var.append(var2)
    var.append(var3)
    
    correlation_matrix = []
    for i in range(3):
        row = []
        for j in range(3):
            #print(i, j)
            corr = calculate_pearson_correlation(data, data[var[i]], data[var[j]])
            row.append(corr)
        correlation_matrix.append(row)

    # Вывод корреляционной матрицы
    print("Корреляционная матрица:")
        
    df = pd.DataFrame(correlation_matrix)

    # Назначение названий столбцов
    df.columns = [var1, var2, var3]

    # Назначение названий строк
    df.index = [var1, var2, var3]

    # Вывод таблицы с названиями столбцов и строк
    print(df)
    print()
```

<h3>2. Коэффициент корреляции Спирмена</h3>

```python
def calculate_spearman_correlation(variable1, variable2):
    # Преобразование переменных в ранги
    ranks1 = get_ranks(variable1)
    ranks2 = get_ranks(variable2)

    # Рассчет разностей рангов
    rank_diffs = [ranks1[i] - ranks2[i] for i in range(len(variable1))]

    # Рассчет суммы квадратов разностей рангов
    rank_diffs_sum_squared = sum([diff ** 2 for diff in rank_diffs])

    # Рассчет коэффициента корреляции Спирмена
    n = len(variable1)
    correlation = 1 - (6 * rank_diffs_sum_squared) / (n * (n ** 2 - 1))

    return correlation

def get_ranks(variable):
    # Получение рангов для переменной
    sorted_variable = sorted(variable)
    ranks = [sorted_variable.index(value) + 1 + (sorted_variable.count(value) - 1) / 2 for value in variable]
    return ranks


# Расчет коэффициента корреляции Спирмена для каждой пары переменных
for combination in combinations:
    var1 = combination[0]
    var2 = combination[1]
    var3 = combination[2]
    
    # Получение данных для соответствующих переменных
    var = []
    var.append(var1)
    var.append(var2)
    var.append(var3)
    
    correlation_matrix = []
    for i in range(3):
        row = []
        for j in range(3):
            #print(i, j)
            corr = calculate_spearman_correlation(data[var[i]], data[var[j]])
            row.append(corr)
        correlation_matrix.append(row)

    # Вывод корреляционной матрицы
    print("Корреляционная матрица:")
        
    df = pd.DataFrame(correlation_matrix)

    # Назначение названий столбцов
    df.columns = [var1, var2, var3]

    # Назначение названий строк
    df.index = [var1, var2, var3]

    # Вывод таблицы с названиями столбцов и строк
    print(df)
    print()
```

<h3>3. Коэффициент корреляции Кендалла</h3>

```python
def calculate_kendall_correlation(variable1, variable2):
    # Преобразование переменных в ранги
    ranks1 = get_ranks(variable1)
    ranks2 = get_ranks(variable2)

    # Рассчет попарных конкордаций
    concordances = [get_concordance(ranks1[i], ranks2[i]) for i in range(len(variable1))]

    # Рассчет суммы попарных конкордаций
    concordance_sum = sum(concordances)

    # Рассчет коэффициента корреляции Кендалла
    n = len(variable1)
    tau = concordance_sum / (0.5 * n * (n - 1))

    return tau

def get_ranks(variable):
    # Получение рангов для переменной
    sorted_variable = sorted(variable)
    ranks = [sorted_variable.index(value) + 1 + (sorted_variable.count(value) - 1) / 2 for value in variable]
    return ranks

def get_concordance(rank1, rank2):
    # Получение попарной конкордации
    if rank1 < rank2:
        return 1
    elif rank1 > rank2:
        return -1
    else:
        return 0

# Расчет коэффициента корреляции Кендалла для каждой пары переменных
for combination in combinations:
    var1 = combination[0]
    var2 = combination[1]
    var3 = combination[2]
    
    # Получение данных для соответствующих переменных
    var = []
    var.append(var1)
    var.append(var2)
    var.append(var3)
    
    correlation_matrix = []
    for i in range(3):
        row = []
        for j in range(3):
            #print(i, j)
            corr = calculate_kendall_correlation(data[var[i]], data[var[j]])
            row.append(corr)
        correlation_matrix.append(row)

    # Вывод корреляционной матрицы
    print("Корреляционная матрица:")
        
    df = pd.DataFrame(correlation_matrix)

    # Назначение названий столбцов
    df.columns = [var1, var2, var3]

    # Назначение названий строк
    df.index = [var1, var2, var3]

    # Вывод таблицы с названиями столбцов и строк
    print(df)
    print()
```
